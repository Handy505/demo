```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def differential(func, x):
    h = 0.1
    return (func(x+h) - func(x))/(x+h-x)

xs = range(0,10)
ys = [f(i) for i in xs]
dys = [differential(f, x) for x in xs]


plt.plot(ys)
plt.plot(dys)
plt.show()
```
