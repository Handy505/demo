```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

class Ball(object):
    def __init__(self):
        self.x = 5
        self.y = 5
        self.accx = 2
        self.accy = 1
        
    def move(self):
        TOP_BOUNDARY = 50
        BUTTOM_BOUNDARY = -50
        RIGHT_BOUNDARY = 50
        LEFT_BOUNDARY = -50
        
        self.x += self.accx 
        if self.x > RIGHT_BOUNDARY:
            self.x = RIGHT_BOUNDARY - (self.x - RIGHT_BOUNDARY)
            self.accx = -self.accx
        elif self.x < LEFT_BOUNDARY:
            self.x = LEFT_BOUNDARY + (self.x - LEFT_BOUNDARY)
            self.accx = -self.accx
            
        self.y += self.accy
        if self.y > TOP_BOUNDARY:
            self.y = TOP_BOUNDARY - (self.y - TOP_BOUNDARY)
            self.accy = -self.accy
        elif self.y < BUTTOM_BOUNDARY:
            self.y = BUTTOM_BOUNDARY + (self.y - BUTTOM_BOUNDARY)
            self.accy = -self.accy
            
b = Ball()    
xpath = []
ypath = []
for _ in range(100):
    b.move()
    print('{},{} ({},{})'.format(b.x, b.y, b.accx, b.accy))
    xpath.append(b.x)
    ypath.append(b.y)
        
    
plt.plot(xpath, ypath, 'ro')
```
