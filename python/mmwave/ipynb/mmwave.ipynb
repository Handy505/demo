```python
import numpy as np


A = 1
times = np.linspace(0, 4, 1000)
freqs = np.linspace(0, 4, 1000)

thedas = 2 * np.pi * freqs * times

def f(x):
    result = A * np.sin(x)
    return result

ys = [f(theda) for theda in thedas]



%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.subplot(311)
plt.plot(times, ys)
plt.subplot(312)
plt.plot(times, freqs)
plt.subplot(313)
plt.plot(times, thedas)
```
